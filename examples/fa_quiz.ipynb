Definire un ASFD che riconosce il linguaggio $L\subseteq\{0,1\}^*$ definito

$L$ Ã¨ quindi l'insieme delle stringhe composte da una sequenza (eventualmente

sequenza di almeno uno 0.

```python
nfa0 = NFA(
    states={'q0', 'q1', 'q2', 'q3'},
    input_symbols={'0', '1'},
    delta={
        'q0': {'0': {'q0'}, '': {'q1'}},
        'q1': {'1': {'q1'}, '': {'q2'}},
        'q2': {'0': {'q3'}},
        'q3': {'0': {'q3'}}
    },
    initial_state='q0',
    final_states={'q3'}
)
```
